```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
from selenium import webdriver
import pandas as pd

```

```python
type_1=[]
name_1=[]
Company_1=[]
Year_1=[]
Salary_1=[]
location_1=[]
Description_1=[]
key_1=[]
Featured_1=[]


for i in range(len(df)):
    driver.get(df['link'][i])
    content = driver.page_source
    soup = BeautifulSoup(content)
    soup
    for a in soup.findAll('article', attrs={'class':'jobTuple bgWhite br4 mb-8'}):
        type_1.append(df['type'][i])
        name=a.find('a', attrs={'class':'title fw500 ellipsis'})
        Company=a.find('a',attrs={'class':"subTitle ellipsis fleft"})
        Year=a.find('span',attrs={'class':"ellipsis fleft fs12 lh16"})
        b=a.find('li',attrs={'class':"fleft grey-text br2 placeHolderLi salary"})
        Salary=b.find('span',attrs={'class':"ellipsis fleft fs12 lh16"})

        d=a.find('li' ,attrs={'class':"fleft grey-text br2 placeHolderLi location"})
        location=d.find('span',attrs={'class':"ellipsis fleft fs12 lh16"})

        Description=a.find('div',attrs={'class':"job-description fs12 grey-text"})
        e=a.find('ul',attrs={'class':"tags has-description"})

        try:
            key_1.append(e.get_text(separator=', '))
        except:
            key_1.append('None')
        try:
            f=a.find('div',attrs={'class':"type br2 fleft grey"})
            Featured=f.find('span',attrs={'class':"fleft fw500"})
            Featured_1.append(Featured.text)
        except:
            Featured_1.append('None')

        Company_1.append(Company.text)
        Year_1.append(Year.text)
        Salary_1.append(Salary.text)
        location_1.append(location.text)
        Description_1.append(Description.text)
        name_1.append(name.text)
        print(i)
```

```python


percentile_list = pd.DataFrame(
    {'Area':type_1,
     'Job_title': name_1,
     'Company': Company_1,
     'Experience':Year_1,
     'Salary':Salary_1,
     'Location': location_1,
     'Description':Description_1,
     'Tag assosiated':key_1,
     'Posting_date':Featured_1  
    })
```
